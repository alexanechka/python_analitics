## 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
df_new=df.drop(columns=['id','count'])
df_corr = df_new.corr()
df_corr
```

```python
plt.figure(figsize=(20,20))

sns.heatmap(df_corr, annot=True, linewidths=1, cmap='coolwarm')

plt.xticks(rotation=30)
plt.title('Матрица корреляции')
```

***По матрице корреляции можно сделать вывод, что на цену больше всего влияют жилая площадь(sqft_living), класс жилья (grade), общая плозадь (sqft_above), жилая площадь 15 ближайших соседей (sqft_living_15)*** 

```python
plt.figure(figsize=(8,6))

plt.bar(df['grade'], df['price'])
plt.title('Зависимость стоимости жилья от класса')
plt.xlabel('Класс')
plt.ylabel('Стоимость')
```

***По графику можно сделать вывод, что чем выше класс жилья, тем выше стоимость, но зависимость нелинейная и на стоимость жилья скорее всего влияет совокупность факторов.***

```python
plt.figure(figsize=(15,12))

sns.barplot(x=df['sqft_above'],y=df['price'])
plt.title('Зависимость стоимости жилья от общей площади (кроме подвала)')
plt.xlabel('Общая площадь')
plt.ylabel('Стоимость')
```

***На графике просматривается зависимость стоимости жилья от общей площади. Зависимость нелинейная.***

```python
plt.figure(figsize=(15,12))

sns.scatterplot(x=df['sqft_living'],y=df['price'])
plt.title('Зависимость стоимости жилья от жилой площади')
plt.xlabel('Жилая площадь')
plt.ylabel('Стоимость')
```

```python
plt.figure(figsize=(15,12))

sns.lineplot(x=df['sqft_living15'],y=df['price'])
sns.lineplot(x=df['sqft_living'],y=df['price'])
plt.title('Зависимость стоимости жилья от жилой площади')
plt.xlabel('Жилая площадь')
plt.ylabel('Стоимость')

plt.legend(['sqft_living15', 'sqft_living'])
```

***Жилая площадь и жилая площадь соседей влияет на цену недвижимости примерно одинаково.***
