```python
df_waterfront = df["waterfront"].value_counts()
df_waterfront.head()
```

```python
plt.figure(figsize=(10, 10))

plt.pie(df_waterfront, labels = df_waterfront.index, autopct='%1.1f%%')

plt.title('Распределение домов от вида на набережную')
```
